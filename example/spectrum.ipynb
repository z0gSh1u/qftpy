```python
#    This example shows how to draw the amplitude (modulus) spectrum of a color image
#    using QFTPy's Quaternion Fourier Transform.
#    
#    by z0gSh1u @ https://github.com/z0gSh1u/qftpy
```

```python
import numpy as np
import quaternion
import matplotlib.pyplot as plt
from PIL import Image

from qftpy.io import im2q
from qftpy.fft import qfft2, qfftshift
```

```python
# The color Lena image comes from http://ndevilla.free.fr/lena/lena.jpg .
Lena = Image.open('lena.jpg')
plt.figure()
plt.imshow(Lena)
```

```python
# Convert Lena image into quaternion matrix
qLena = im2q(Lena)
qLena.dtype
```

```python
# Perform 2D Quaternion Fourier Transform
spec = qfftshift(qfft2(qLena))
```

```python
# Display.
plt.figure()
plt.imshow(np.log(np.abs(spec)), cmap='gray')
```
